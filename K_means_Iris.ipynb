## K-means

### Importação das bibliotecas

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
%matplotlib inline
from sklearn import datasets
```

### Carregando a base de dados [Iris](https://archive.ics.uci.edu/ml/datasets/iris)

```python
#Iris Dataset
iris = datasets.load_iris()
X = iris.data
```

```python
iris.target
```

```python
iris.target_names
```

```python
X
```

### Instanciando o classificador não supervisionado [K-Means](https://scikit-learn.org/stable/modules/clustering.html#k-means)

```python
#KMeans
km = KMeans(n_clusters=3)
```

```python
km.fit(X)
```

```python
km.predict([[1.1, 3.2, 0.9, 4.4]])
```

```python
km.predict(X)
labels = km.labels_
```

```python
labels
```

### Plotando um gráfico para mostrar o resultado

```python
#Plotting
fig = plt.figure(1, figsize=(7,7))
ax = Axes3D(fig, rect=[0, 0, 0.95, 1], elev=40, azim=135)
ax.scatter(X[:, 3], X[:, 0], X[:, 2],
          c=labels.astype(np.float), edgecolor="k", s=50)
ax.set_xlabel("Petal width")
ax.set_ylabel("Sepal length")
ax.set_zlabel("Petal length")
plt.title("K Means", fontsize=14)
```
